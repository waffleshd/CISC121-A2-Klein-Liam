```python
'''
I checked this solution on LeetCode: https://leetcode.com/problems/convert-date-to-binary/solutions/7221141/convert-date-to-binary-easy-optimal-solution-mathematical-approach-c-java-python/
And it did pretty much exactly what I was going to do.
'''

def toBin(self,date):
    date = int(date)
    remainders = "" # string to store remainders
    while date != 0:
        remainders += str(date % 2) # get remainder when divided by 2
        date = date // 2
    return remainders[::-1] # reverse the string to get the correct binary

def convertDateToBinary(self, date):
    year = self.toBin(date[:4]) #call toBin() for year, month, and day
    month = self.toBin(date[6:7])
    day = self.toBin(date[8:10])
    return (year + "-" + month + "-" + day) # construct and return the final binary string
```

```python
'''
After reviewing this solution on LeetCode: https://leetcode.com/problems/number-of-1-bits/solutions/3218849/191-solution-step-by-step-explanation/

I realized I was overcomplicating things. We can use the AND operator
to check if the last bit of n is a 1. If it returns 1, we increment our counter by 1.

After this, no matter what, we shift n to the right, so we can inspect the next bit
in n.
'''

def hammingWeight(self, n):
    counter = 0
    while n != 0: # if n == 0, we return our count
        if (n & 1) == 1: counter += 1   # if the final bit of n is 1, we increment the counter by 1
        n >>= 1 # shift n by one, so that the next bit can be inspected
    return counter
```

**Problem Decomposition**

We are given an integer x, and must reverse its digits. However, if the reversed integer becomes larger than the 32 bit limit, we simply return 0.

The difficulty arises with that we cannot use 64 bit integers at all. This means we cannot simply check if a given number is larger than the 32 bit limit.

**Pattern Recognition**

We can use a recursive function to reverse x after it has been converted into a string.

I first thought there may be a trick to use binary and reverse the order there, but I could not find any patterns.

**Abstraction**

When looking at each element in the integer, we can disregard any leading 0's, however we must make sure to save the negative if it exists.

**Algorithmic Thinking**

Input - Integer x

Output- Integer reversed_x

Constraints:
- -231 <= x <= 231 - 1

image.png

The recursive function should make sure we ignore any leading zeros. These are unnecessary to store. Additionally, if x is negative, the function should recognize this, and add it to the front of the reversed number.

The base case should determine if the string x's length is 0, then the function should stop. If x is negative, we can also stop when we find the - sign.
The recursive steps should check for leading 0s, check for - signs, and otherwise simply call itself again, to reverse the number.

The call stack will organize the calls such that the 'last' call will be ran first (tail recursion).

We are using tail recursion in order to deal with the last digit of x first, so we can add it to the reversed_x string.

The subproblems can be combined seamlessly with an if else statement or a switch statement.

```python
'''
Draft code.

I looked at the top solutions on LeetCode, and none of them used recursion.
It would be very simple to solve this problem with some simple string slicing,
but since this is meant as practice for recursion, here we go:

I found this solution on LeetCode: https://leetcode.com/problems/reverse-integer/solutions/7199544/c-python-digit-pop-push-string-reverse-beats-100-visuals-o-n-time-o-1-space/
to give me the idea of ignoring the sign of x until the end.

Everything else is really just some basic string slicing and some tail recursion
to get the reversed integer.
'''

def reverse(self, x):
    if x < 0: # if x is negative, we set the sign to be -1, and set x to be positive
        sign = -1
        x = abs(x)
    else: sign = 1 # else, we leave it be
    if len(str(x)) > 1: # if we still have more than one digit left...
        last_digit = str(x)[-1] # get the last digit
        x = str(x)[:-1] # get x without the last digit
        return int(last_digit + str(self.reverse(int(x)))) * sign # recursive call and multiply reversed by the sign at the end
    return x
```
